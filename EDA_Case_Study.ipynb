**Univariate Analysis of categorical values between defaulters vs non-defaulters**

```python
#define a function to plot the chart multiple times to analyze the categorical values

def  plotunivariatecatfunc(col):
    
    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize = (20,6))
    
    sns.countplot(x = col, data = df0, ax = ax1)
    ax1.set_ylabel('Count in number')
    ax1.set_title(f'Dissemination of {col} for Non-Defaulters', fontsize = 14)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=40, ha="right")
    
    for percentage in ax1.patches:
        ax1.annotate('{:.1f}%'.format((percentage.get_height()/len(df0))*100), (percentage.get_x()+0.1, percentage.get_height()+50))
        
    sns.countplot(x = col, data = df1, ax = ax2)
    ax2.set_ylabel('Count in number')
    ax2.set_title(f'Dissemination of {col} for Defaulters', fontsize = 14)
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=40, ha="right")
                  
    for percentage in ax2.patches:
        ax2.annotate('{:.1f}%'.format((percentage.get_height()/len(df1))*100), (percentage.get_x()+0.1, percentage.get_height()+50))
    
    plt.show()
```

```python
def univariateplotfunc(df,col,title,hue =None):
    
    sns.set_style('darkgrid')
    sns.set_context('poster')
    plt.rcParams["axes.labelsize"] = 15
    plt.rcParams['axes.titlesize'] = 15
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue, palette='bright') 
    plt.setp(ax.get_xticklabels(), rotation = 45, ha= "right", rotation_mode = "anchor")
    plt.show()
```

**Continuous Variable Analysis**

```python
def plotunivariatecontfunc(col):
    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize = (15,5))
    
    sns.distplot(a = df0[col], ax = ax1)
    ax1.set_title(f'Dissemination of {col} for Non-Defaulters', fontsize = 12)
    
    sns.distplot(a = df1[col], ax = ax2)
    ax2.set_title(f'Dissemination of {col} for Defaulters', fontsize = 12)
                  
    plt.show()
```
